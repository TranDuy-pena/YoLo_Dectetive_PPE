```python
import cv2
from ultralytics import YOLO

# Load mô hình YOLO đã huấn luyện
model = YOLO('runs/detect/train63/weights/best.pt')  # Đảm bảo đường dẫn đúng

# Mở video (có thể là file hoặc stream từ camera)
cap = cv2.VideoCapture('D:\AnToan\\AnToanCongTrinh\\ThuMoHinh\\test1.mp4')  # Đổi 'video_path.mp4' thành đường dẫn video của bạn

if not cap.isOpened():
    print("Error: Không thể mở video.")
    exit()

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Dự đoán với mô hình YOLO
    results = model(frame)  # Dự đoán trên từng frame của video

    # Vẽ bounding boxes lên frame
    frame = results[0].plot()  # vẽ bounding boxes lên ảnh

    # Hiển thị kết quả
    cv2.imshow("YOLO Detection", frame)

    # Để dừng, nhấn phím 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Giải phóng tài nguyên
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO

# Load mô hình
model = YOLO('D:\\AnToan\\AnToanCongTrinh\\runs\\detect\\train63\\weights\\best.pt')  # đổi đường dẫn model của bạn

# Mở camera (camera mặc định id = 0)
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    # Dự đoán trên frame
    results = model(frame)

    # Vẽ kết quả lên frame
    annotated_frame = results[0].plot()

    # Hiển thị frame
    cv2.imshow('PPE Detection', annotated_frame)

    # Ấn 'q' để thoát
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Giải phóng
cap.release()
cv2.destroyAllWindows()
```
